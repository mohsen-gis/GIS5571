```python
import arcgis as arc
from arcgis.geoanalytics import use_proximity
import arcgis 

# dataset = '/arcgis/stpaulNet/stpaul.shp'
# Set the tool environment settings
arcgis.env.verbose = True
arcgis.env.defaultAggregations = True
my_layer = arcgis.features.FeatureLayer('https://smartcommunityhealth.ahc.umn.edu/COVID/ows?service=WFS&version=1.0.0&request=GetFeature&typeName=COVID%3Acarlson_final&propertyName=location,date,region_id,variable,value,variable_id,region&outputFormat=application%2Fjson&viewparams=id:751025')


# Run the tool Create Buffers 
output = use_proximity.create_buffers(input_layer = my_layer, distance = 10, distance_unit = 'Meters' , method = "Geodesic", dissolve_option = "All", output_name = "bufferred_10m_JnOnline")

```
